Question 4-Write a python function to perform selection sort on a given string.

```python
lst= list('azac')#try doing with recursion
n=len(lst)

for i in range(n):
    min=i
    for j in range(i+1,n):
        if(lst[j]<lst[min]):
            temp=lst[j]
            lst[j]=lst[min]
            lst[min]=temp
print(lst)
```

Question 5- Find the factorial of a given number using recursion.

```python
def fact(n):
    if(n==0):
        return 1
    else:
        return n*fact(n-1)
m=int(input("enter number: "))
print(fact(m))
```

Question 7-Write a program to print the Fibonacci Sequence till n-values where n is user input

```python
n=int(input("Enter n: "))
a=0
b=1 
print(a, ",", b)
for i in range(2,n+1):
    c=a+b
    print(c, end=", ")
    a=b
    b=c
```
